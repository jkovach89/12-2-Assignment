```python
plt.plot(edu_df[edu_df.STATE == 'CALIFORNIA']['TOTAL_REVENUE'], label='Revenue')
plt.plot(edu_df[edu_df.STATE == 'CALIFORNIA']['TOTAL_EXPENDITURE'], label='Expenditure')
plt.legend()
plt.title('California: Revenue vs. Expenditure')
plt.show()
```

```python
def strip_null(series):
    return series.dropna(how='all')

am4 = strip_null(edu_df[edu_df.STATE == 'CALIFORNIA']['AVG_MATH_4_SCORE'])
am8 = strip_null(edu_df[edu_df.STATE == 'CALIFORNIA']['AVG_MATH_8_SCORE'])
ar4 = strip_null(edu_df[edu_df.STATE == 'CALIFORNIA']['AVG_READING_4_SCORE'])
ar8 = strip_null(edu_df[edu_df.STATE == 'CALIFORNIA']['AVG_MATH_4_SCORE'])


plt.figure(figsize=(15,15))

plt.subplot(3,2,1)
plt.plot(am4, label='Math')
plt.plot(ar4, label='Reading')
plt.legend()
plt.title('California: 4th Grade Math and Reading Scores')

plt.subplot(3,2,2)
plt.plot(am8, label='Math')
plt.plot(ar8, label='Reading')
plt.legend()
plt.title('California: 8th Grade Math and Reading Scores')

plt.subplot(3,2,3)
plt.hist(am4)
plt.title('California: Distribution of 4th Grade Math Scores')

plt.subplot(3,2,4)
plt.hist(ar4)
plt.title('California: Distribution of 4th Grade Reading Scores')

plt.subplot(3,2,5)
plt.hist(am8)
plt.title('California: Distribution of 8th Grade Math Scores')

plt.subplot(3,2,6)
plt.hist(ar8)
plt.title('California: Distribution of 8th Grade Reading Scores')
plt.show()
```
