```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
%matplotlib inline
data = pd.read_csv('/Users/JKovach/Downloads/APCspend.csv', encoding = 'ISO-8859-1')
print(data)
```

```python
data['clean_title'] = data['Journal title'].str.lower()
#data['Journal title'].value_counts()
data['clean_title'].value_counts()
```

```python
data['cost_lb'] = data['COST (£) charged to Wellcome (inc VAT when charged)'].str.replace('£','')
data['cost_num'] = data['cost_lb'].str.replace('$','')
data['cost_float'] = data['cost_num'].apply(lambda x: float(x))
data_clean = data[data.cost_float < 20000]
print(np.mean(data_clean['cost_float']))
```
