```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
%matplotlib inline
print('Import Complete')
```

```python
traf = pd.read_csv('/Users/Jkovach/Downloads/los-angeles-traffic-collision-data/traffic-collision-data-from-2010-to-present.csv')
print(traf.head(n=5))
```

```python
#Showing the information in a seaborn facet grid
sns.set(style='ticks') #Preset aesthetic

#Tell seaborn about data structure
g = sns.FacetGrid(traf, row='Crime Code', col='Victim Sex')
#Name the plot type and the variable to be plotted using the created structure
g.map(plt.hist, 'Time Occurred', color='steelblue', lw=0, bins=20)
#Create space for titles
plt.subplots_adjust(top=.85)
#Title informative axis name
g.set_axis_labels('Time of Incident')
plt.suptitle('Traffic Incidents by Time and Victim Gender')
#Remove excess lines around plot
sns.despine(trim=False)
plt.show()
```

```python
# Comparing groups using boxplots.
ax = sns.boxplot(x='Time Occurred',y='Area Name',hue='Crime Code',data=traf)  
plt.title('Traffic Incidents by Area and Time of Day')
sns.despine(offset=10, trim=True)
ax.set(xlabel='Time of Day', ylabel='Area')
plt.show()
```

```python
g = sns.lmplot(y='Time Occurred', # Variable 1.
               x='Area Name', # Variable 2.
               data=traf, # Data
               fit_reg=True, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.4}) # Set points to semi-transparent to see overlapping points.
g.set_ylabels("Time of Incident")
g.set_xlabels("Date")
plt.title('Scatterplot: Meal Tips by Total Bill')
plt.show()
```
